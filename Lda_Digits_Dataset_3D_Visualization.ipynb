```python
# steps
# 1. data
# 2. seperate them in independent and dependent data
# 3. split the data into training and testing bcz Lda is supervised so first sliting
# 4. Standardization
# 5. preprocessing
X=df.iloc[:,:-1] # Independent
y=df['target']
```

```python
X
```

```python
# standerdiztion
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
```

```python
X_std=sc.fit_transform(X)
#df_scaled = pd.DataFrame(sc.fit_transform(X), columns=df['pixel_0_0	pixel_0_1	pixel_0_2	pixel_0_3	pixel_0_4	pixel_0_5	pixel_0_6	pixel_0_7	pixel_1_0	pixel_1_1	...	pixel_6_6	pixel_6_7	pixel_7_0	pixel_7_1	pixel_7_2	pixel_7_3	pixel_7_4	pixel_7_5	pixel_7_6	pixel_7_7'])
```

```python
X_std.shape
```

```python
y
```

```python
# Splitting tahe dataon training and testing
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_std,y, test_size=0.3,random_state=45)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
# Implementation Of LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
df['target'].nunique()
```

```python
lda=LinearDiscriminantAnalysis(n_components=9)# n_components = n-1, n is the no of clases in the target column
```

```python
x_train=lda.fit_transform(x_train,y_train)
```

```python
x_train.shape
```

```python
x_test=lda.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import *
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# 3D LDA Scatter Plot

import plotly.express as px
fig = px.scatter_3d(df, x=x_train[:,8], y=y_train, z=x_train[:,7],
                    labels={
                        "x": "LD1",
                        "y": "LD2",
                        "z": "LD3"
                    },
                    opacity=1, color = y_train)


fig.update_layout(scene=dict(xaxis_showgrid=True, xaxis_gridwidth=1, xaxis_gridcolor='lightgrey',
                             xaxis_zeroline=True, xaxis_zerolinewidth=1, xaxis_zerolinecolor='lightgrey',
                             xaxis_showline=True, xaxis_linewidth=1, xaxis_linecolor='black',
                             yaxis_showgrid=True, yaxis_gridwidth=1, yaxis_gridcolor='lightgrey',
                             yaxis_zeroline=True, yaxis_zerolinewidth=1, yaxis_zerolinecolor='lightgrey',
                             yaxis_showline=True, yaxis_linewidth=1, yaxis_linecolor='black',
                             zaxis_showgrid=True, zaxis_gridwidth=1, zaxis_gridcolor='lightgrey',
                             zaxis_zeroline=True, zaxis_zerolinewidth=1, zaxis_zerolinecolor='lightgrey',
                             zaxis_showline=True, zaxis_linewidth=1, zaxis_linecolor='black'))


# Change chart background color
fig.update_layout(scene=dict(xaxis_backgroundcolor="white",
                             yaxis_backgroundcolor="white",
                             zaxis_backgroundcolor="white"))

# Set figure title
fig.update_layout(title_text="3D LDA scatter plot")

# Update marker size
fig.update_traces(marker=dict(size=3))

fig.show()
```
